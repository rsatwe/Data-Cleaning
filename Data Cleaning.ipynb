## IQR (Interquartile range ) technique for outlier treatment

```python
def outlier_treatment(col):
    
    sorted(col)
    Q1,Q3 = np.percentile(col , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
lower_range,upper_range = outlier_treatment(diamond_df['price'])
print("Lower Range:",lower_range)
print("Upper Range:",upper_range)
```

```python
lower_diamond_df = diamond_df[diamond_df['price'].values < lower_range]
lower_diamond_df
```

```python
upper_diamond_df = diamond_df[diamond_df['price'].values > upper_range]
upper_diamond_df
```

```python

lower_outliers = lower_diamond_df.value_counts().sum(axis=0)
upper_outliers = upper_diamond_df.value_counts().sum(axis=0)
total_outliers = lower_outliers + upper_outliers
print("Total Number of Outliers:",total_outliers)
```

Let us list down the row numbers that contain outliers:

```python
lower_index = list(diamond_df[ diamond_df['price'] < lower_range ].index)
upper_index = list(diamond_df[ diamond_df['price'] > upper_range ].index)
total_index = list(lower_index + upper_index)
print(total_index)
```

## Drop the outlier rows

```python
print("Shape Before Dropping Outlier Rows:", diamond_df.shape)
diamond_df.drop(total_index, inplace = True)
print("Shape After Dropping Outlier Rows:", diamond_df.shape)
```
